```python
dna1 = 'ATGAGTCTCT'
dna2 = 'CTGTCTCCTG'
m = len(dna1)
n = len(dna2)
d = 10
e = 0.5
path = {}
smatrix = [[0 for _ in range(m+1)] for _ in range(n+1)]
amatrix = [[0 for _ in range(m)] for _ in range(n+1)]
bmatrix = [[0 for _ in range(m+1)] for _ in range(n)]

# задаем начальные условия для матриц

for i in range (1, m+1):
    smatrix[0][i]= -d-e*(i-1)
for i in range (1, n+1):
    smatrix[i][0]= -d-e*(i-1)
for i in range (0, m):
    amatrix[0][i] = smatrix[0][i+1] - d
for i in range (0, n):
    bmatrix[i][0] = smatrix[i+1][0] - d

for i in range (1, m+1):
    for j in range (1, n+1):
        if j < n and i < m:
            a = amatrix[i-1][j] - e
            b = smatrix[i-1][j] - d
            amatrix[i][j] = max(a, b)
            c = bmatrix[i][j-1] - e
            d = smatrix[i][j-1] - d
            bmatrix[i][j] = max(c, d)
        
        if dna1[j-1]==dna2[i-1]:
            s = 5
        else:
            s = -4
        right = amatrix[i][j-1]
        down = bmatrix[i-1][j]
        diag = smatrix[i-1][j-1] + s
        if diag == down or diag == right:
            diag -=1
        smatrix[i][j]=max(right, down, diag) # условия нужно вписать
        ij = str(i)+str(j)
        if max(right, down, diag) == right:
            path[ij]= 'right'
        elif max(right, down, diag) == down:
            path[ij] = 'down'
        else:
            path[ij] = 'diag'
score = smatrix[m][n]
print(round(score, 1))

print(*smatrix, sep='\n')
print(*amatrix, sep='\n')
```
