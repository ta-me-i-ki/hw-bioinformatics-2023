```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
     

df = pd.read_csv("BRCA_pam50.tsv", sep="\t", index_col=0)
sample = ['ESR1', 'PGR', 'ERBB2', 'MKI67']
expr = []
gene = []
sub = []
for i in sample:
    expr = expr + list(df[i])
    sub = sub + list(df['Subtype'])
    for j in range(len(df[i])):
        gene.append(i)

tab = pd.DataFrame({
    'Expression': expr,
    'Gene': gene,
    'Subtype': sub
})
print(tab)

fig = plt.gcf()
fig.set_size_inches(20, 10)
[plt.axvline(x, color = 'black', linestyle='--') for x in [i + 0.5 for i in range(5)]]
graph = sns.boxplot(x = 'Subtype', y = 'Expression', hue = 'Gene', data = tab)
#luminal схожи между собой и наибольшим образом отвечают за экспрессию генов esr1, pgr.
#у triple-negative проблема со экспрессией всего, кроме mki67, он схож с остальными подтипами (даже выше)
#her2-enriched показывает высокую экспрессию гена erbb2, средние показатели mki67, низкая экспрессия espr1,pgr 
#здоровый образец различается с нормой, но это можно связать с плохим средним здоровьем у людей? хз... mki67 просто сильно различается
```
