```python

from Bio import SeqIO
from itertools import product

sequence = ''
for record in SeqIO.parse("SARS-CoV-2.fasta", "fasta"):
    sequence += record.seq


k = int(input()) 
dic_kmers = {}
for i in product('ATGC', repeat = k):
    dic_kmers[''.join(i)] = []

start = 0
while True:
    if start == len(sequence) - k + 1:
        break
    kmer = sequence[start:start + k]
    if kmer in dic_kmers: # это чтобы если там вдруг какие-нибудь другие символы от плохо прочитанных нуклеотидов мы их не записывали
        # но если вдруг хотим записать то можно убрать условие и будут вообще все варианты
        dic_kmers[kmer]=dic_kmers[kmer]+[start + 1]
    start +=1
print(dic_kmers)
```

```python
mass = list(map(int, input().split(', ')))
dct = {}
sums = set()
pref = 0
for i in range (0, len(mass)):
    pref += mass[i]
    if mass[i] == 0:
        print(i)
    if pref == 0:
        print (0, i, sep=':')
    else:
        if pref in sums:
            print(dct[pref] + 1, i, sep = ':')
        else:
            sums.add(pref)
            dct[pref] = i
```

```python
from Bio import SeqIO
from itertools import product

sequence = ''
for record in SeqIO.parse("SARS-CoV-2.fasta", "fasta"):
    sequence += str(record.seq) 

def kmers (k, sequence):
    dic_kmers = {}
    start = 0
    while True:
        if start == len(sequence) - k + 1:
            break
        kmer = sequence[start:start + k]
        dic_kmers[kmer]=dic_kmers.get((kmer),[])+[start + 1]
        start +=1
    return dic_kmers

k = 0

while True:
    k +=1
    dic_kmers = kmers (k, sequence)
    m, keys = len(dic_kmers), 0 
    for i in dic_kmers: 
        if len(dic_kmers[i]) <= 1: 
            keys += 1 
        else: 
            break      
    if m == keys: 
        print(k) 
        break
```
